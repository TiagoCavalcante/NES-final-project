```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/filtered_data.csv")
```

```python
df
```

```python
# Análise de Correlação entre Pontuações do SAT e Taxas de Admissão

sat_data = df[['ADM_RATE', 'SATVRMID']].dropna()
plt.figure(figsize=(8, 6), facecolor='#A4B8C4')
plt.scatter(sat_data['SATVRMID'], sat_data['ADM_RATE'], alpha=0.5, color='purple')
plt.gca().set_facecolor('#A4B8C4')
plt.title('Correlação entre Pontuações do SAT e Taxas de Admissão', fontsize=14)
plt.xlabel('Pontuação Média do SAT (Leitura)', fontsize=12, color='purple')
plt.ylabel('Taxa de Admissão', fontsize=12, color='purple')
plt.grid(True, color='black')
plt.show()
```

```python
# Análise de Comparação de Custo e Renda Familiar Média

cost_income_data = df[['COSTT4_A', 'FAMINC']].dropna()
plt.figure(figsize=(8, 6), facecolor='#0A2538')
plt.scatter(cost_income_data['FAMINC'], cost_income_data['COSTT4_A'], alpha=0.5, color='orange')
plt.title('Comparação de Custo e Renda Familiar Média', fontsize=14, color='orange')
plt.xlabel('Renda Familiar Média (USD)', fontsize=12, color='orange')
plt.ylabel('Custo Anual da Universidade (USD)', fontsize=12, color='orange')
plt.yticks(color='orange')
plt.show()
```

```python
# Ranking das Cidades Mais Populares

city_data = df['CITY'].value_counts().head(10)
plt.figure(figsize=(8, 6), facecolor='gray')
plt.gca().set_facecolor('#A4B8C4')
city_data.plot(kind='bar', color='teal')
plt.title('Cidades Mais Populares para Estudo',fontsize=14)
plt.xlabel('Cidade', fontsize=12)
plt.ylabel('Número de Universidades', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Análise da Distribuição Racial e Étnica nas Universidades

colunas = ['UGDS_WHITE', 'UGDS_BLACK', 'UGDS_HISP', 'UGDS_ASIAN', 'UGDS_AIAN', 'UGDS_NHPI', 'UGDS_2MOR', 'UGDS_NRA', 'UGDS_UNKN']
racas = ['Brancos', 'Negros', 'Hispânicos', 'Asiáticos', 'Indígenas', 'Nativos do Pacífico', 'Dois ou mais grupos', 'Estrangeiros', 'Desconhecido']

plt.figure(figsize=(8, 6), facecolor='#0A2538')
df[colunas].mean().plot(kind='bar', color=['lightblue', 'lightgreen', 'lightcoral', 'yellow', 'lightpink', 'lightsalmon', 'lightgrey', 'cyan', 'lightgoldenrodyellow'])
plt.xticks(ticks=range(len(racas)), labels=racas, rotation=45, color='lightblue')
plt.yticks(color='lightblue')
plt.title('Distribuição Racial e Étnica nas Universidades', fontsize=14, color='lightblue')
plt.xlabel('Grupos Racial/Étnicos', fontsize=12, color='lightblue')
plt.ylabel('Porcentagem Média (%)', fontsize=12, color='lightblue')
plt.grid(True)
plt.show()
```

```python
# Distribuição de Matrículas por Faixa Etária em Universidades Públicas

plt.figure(figsize=(8, 6), facecolor='#A4B8C4')
df[['NPT4_048_PUB', 'NPT4_3075_PUB', 'NPT4_75UP_PUB']].mean().plot(kind='bar', stacked=True, color='purple')
plt.gca().set_facecolor('#A4B8C4')
plt.title('Distribuição de Matrículas por Faixa Etária em Universidades Públicas', fontsize=14)
plt.ylabel('Número Médio de Matrículas', fontsize=12)
plt.grid(True)
plt.show()
```
